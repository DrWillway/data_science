```python
# question 1
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-4 * np.sin(2 * np.pi * x))

# Generate 1000 evenly spaced x values from 0 to 1
x_values = np.linspace(0, 1, 1000)

# Compute f(x) for each x
y_values = f(x_values)

# Create a plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='f(x) = e^(-4*sin(2*pi*x))', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = e^(-4*sin(2*pi*x))')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# question 3
# part 1
import numpy as np

# Calculate the function values f(x) as before
def f(x):
    return np.exp(-4 * np.sin(2 * np.pi * x))

# Generate 1000 evenly spaced x values from 0 to 1
x_values = np.linspace(0, 1, 1000)

# Calculate delta x
delta_x = x_values[1] - x_values[0]

# Compute the function values f(x)
y_values = f(x_values)

# Compute the differences ∆f using np.diff
delta_f = np.diff(y_values) / delta_x

# part 2
import numpy as np
import matplotlib.pyplot as plt

# Calculate the adjusted x values
x_values_adjusted = x_values[:-1]

# Compute the derivative f'
derivative = delta_f / delta_x

# Create a plot
plt.figure(figsize=(8, 6))
plt.plot(x_values_adjusted, derivative, label="f'(x)", color='b')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("Plot of Derivative f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.array([1,3,5,6])
```

```python
l = [[1,2,3],[10,44,55]]
```

```python
x = np.array([[1,2,3],[10,44,55]])
x
```

```python
x[0,1]
```

```python
x
```

```python
x.T[1]
```

# Boolean Operations

Some of the most useful features of the numpy array is boolean indexing. For example, suppose I create an array of evenly spaced values between 1 and 4

```python
x = np.linspace(1, 4, 100)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

The following evaluates $x>2$ for all points in the `x` array

```python
x
```

```python
x>2
```

I can then use this array to index elements of $x$

```python
y[x>2]
```

I can even use the `x>2` array to index `y` and plot a subset of each array

```python
plt.plot(x,y)
plt.plot(x[y<0.5], y[y<0.5])
```

You can also combine multiple boolean conditions together:

```python
arr =(x>2) * (x<3)
```

You can think of `True`s as being 1s and `False`s as being 0s; thats how the mathematical operations are achieved on boolean values. Finally you can index with these as well

```python
plt.plot(x,y)
plt.plot(x[(y>-0.5) * (y<0.5)], y[(y>-0.5) * (y<0.5)])
```
